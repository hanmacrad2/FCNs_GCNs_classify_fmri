### Classification Results breakdown across time
- Proportion of correct predictions on test set across time

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Plot params
MEDIUM_SIZE = 14
BIGGER_SIZE = 17
plt.rc('font', size=MEDIUM_SIZE);
plt.rc('axes', titlesize=MEDIUM_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

### 1. Plot proportion of correctly classified test cases correct across Blocks

```python
def plot_proportion(prop_correct, data_type, model_type):
    'Plot proportion of correctly classified test cases correct across Blocks'

    #Block index
    index  = np.arange(0, len(prop_correct))

    #Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(index, 100*prop_correct, color = 'r') # , type = '.') #, figsize = (9,7))
    plt.xlabel('Block')
    plt.ylabel('% Subjects correct (109 in test set)')
    plt.title(f'{model_type} on {data_type} data - proportion correct vs time')
    plt.show()
```
